# A exersise with lists

**Cineplus**

The owners of the CinePlus theater want me to develop software for ticket and food sales, so that it can store all
ticket, popcorn, and soda sales made at each function in a dictionary list.
They have 4 types of customers
* Normal
* Platinum
* Gold
* Diamond

The Normal customer does not have any discount on their purchases.

The Platinum customer has a 5% discount on tickets and a 10% discount on popcorn and soda.

The Gold customer has a 20% discount on tickets, a 30% discount on popcorn, and a 40% discount on soda.

The Diamond customer has a 50% discount on tickets, a 40% discount on popcorn, and a 50% discount on soda.

The application must start with an empty list. When a customer comes to make a purchase, he is asked for the data, type of customer, number of receipts, number of popcorn, number of sodas. Based on this data, the total to be paid must be calculated. This data, obtained and calculated, must be stored in a dictionary that in turn must be stored in the list.

The program must present the following menu

1. New sale.
2. Show all sales
3. Show total of receipts, popcorn and sodas sold.
4. Show the number of customers in each category
5. Show the total sales
6. Exit

discount = $\displaystyle\frac{x\times value}{100}$

```python
#---Â¡PRICES-FUNCTION-INCLUIDS-THE-DISCOUNT!
def Prices(object_price, percent):
    return object_price - ((percent * object_price)/100)

def customer_type_list():
    return ['Normal', 'Platinum', 'Gold', 'Diamond']

#---Â¡USTOMER-INFORMATION-FUNCITON!
def Customer_information():
    
    #--Â¡TO-ADD-CUSTOMER-INFORMATION!
    customer_type = input(
        """
        Add customer type; 
        Normal; if you are a regular customer, 
        Platinum; if you are a platinum customer, 
        Gold; if you are a gold customer,
        Diamond; if you are a diamond customer.
        """)
    number_ticks = int(input('Add number of ticks; '))
    number_popcorns = int(input('Add number of popcorns; '))
    number_drinks = int(input('Add number of drinks; '))
    ticks_price = float(int(input('Add ticks price; ')))
    popcorns_price = float(input('Add popcorn price; ' ))
    drink_price = float(input('Add drinks price; ')) 

    #---Â¡SELECT-THE-CUSTOMER-TYPE!
    if customer_type == 'Normal':
        ticks_price = number_ticks * ticks_price
        popcorns_price = number_popcorns * popcorns_price
        drink_price = number_drinks * drink_price
    elif customer_type == 'Platinum':
        ticks_price = number_ticks * Prices(ticks_price, 5, )
        popcorns_price = number_popcorns * Prices(popcorns_price, 10)
        drink_price = drink_price * Prices(drink_price, 10)
    elif customer_type == 'Gold':
        ticks_price = number_ticks * Prices(ticks_price, 30)
        popcorns_price = number_popcorns * Prices(popcorns_price, 40)
        drink_price = drink_price * Prices(drink_price, 50)
    elif customer_type == 'Diamond':
        ticks_price = number_ticks * Prices(ticks_price, 50)
        popcorns_price = number_popcorns * Prices(popcorns_price, 40)
        drink_price = drink_price * Prices(drink_price, 50)
    #---Â¡ALOWW-SENTENCE-IS-CREATED-FOR-A-ERROR-TYPE-CLIENT!
    else:
        number_ticks = 'Error'
        number_popcorns = 'Error'
        number_drinks = 'Error'
        ticks_price = 'Error'
        popcorns_price = 'Error'
        drink_price = 'Error' 
        
    #---Â¡TOTAL-SALES-AND-TOTAL-PRICE!
    total_sales = number_ticks + number_popcorns + number_drinks
    total_pay = ticks_price + popcorns_price + drink_price

    #---Â¡CUSTOMER-DICTIONARY!
    customer_dic = {
        'Customer_type' : customer_type,
        'Number_ticks' : number_ticks,
        'Number_popcorns' : number_popcorns,
        'Number_drinks': number_drinks,
        'Ticks_price' : ticks_price,
        'Popcorns_price' : popcorns_price,
        'Drink_price' : drink_price,
        'Total_sales' : total_sales,
        'Total_pay' : total_pay
    }
    return customer_dic

#---Â¡CUSTOMER-LIST!
customer_list = []

#---Â¡MAIN!
while True:
    #--Â¡PRINT-THE-MENU-INFORMATION!
    customer_tipe_inf = customer_type_list()
    print('The customer types in this cinema is:\n', customer_tipe_inf, '\n \n Below is the cinema menu')
    print(
        """
        1. New sale.
        2. Show all sales
        3. Show total of receipts, popcorn and sodas sold.
        4. Show the number of customers in each category
        5. Show the total sales
        6. Exit
        """
    )

    #---Â¡ADD-DESIRED-OPTIONS!
    option = input('Add desired options; ')

    #---Â¡MENU-OPTIONS-SECUENCE!
    if option == '6':
        break
    elif option == '1':
        call_customers = Customer_information()
        customer_list.append(call_customers)
        print(customer_list)
        #---Â¡WE-REMOVED-THE-ERRORS-VALUES!
        for customer in customer_list:
            if type(customer['Number_ticks']) == type('str'):
                customer_list.remove(customer)
                print('Â¡WARING; YOU ENTERED AN INVALID CLIENT! \n')
    elif option == '2':
        for customer in customer_list:
            print(
                'Customer_type; ', customer['Customer_type'], '.', 'Number_ticks; ', customer['Number_ticks'], '.',
                'Number_popcorns; ', customer['Number_popcorns'], '.', 'Total_pay; ', customer['Total_pay'], '\n'
            )
    elif option == '3':
        total_ticks = 0
        total_popcorns = 0
        total_drinks = 0
        for total in customer_list:
            total_ticks += total['Number_ticks']
            total_popcorns += total['Number_popcorns']
            total_drinks += total['Number_drinks']
        print('The total of ticks sales is; ', total_ticks,'.', 'The total of popcorns sales is; ', 
                      total_popcorns,'.', 'and the total of drinks sales is; ', total_drinks, '\n')
    elif option == '4':
        customer_normal = 0
        customer_platinum = 0
        customer_gold = 0
        customer_diamond = 0
        for customer in customer_list:
            if customer['Customer_type'] == 'Normal':
                customer_normal += 1
            elif customer['Customer_type'] == 'Platinum':
                customer_platinum += 1
            elif customer['Customer_type'] == 'Gold':
                customer_gold += 1
            elif customer['Customer_type'] == 'Diamond':
                customer_diamond += 1
        print('The total Normal Customers is; ', customer_normal, '\n The total Platinum Customers is; ', customer_platinum, 
             '\n The total Gold Customers is; ', customer_gold, '\n The total Diamond Customers is; ', customer_diamond, '\n')
    elif option == '5':
        Total_sales = 0
        Total_pay = 0
        for customer in customer_list:
            Total_sales += customer['Total_sales']
            Total_pay += customer['Total_pay']
        print('The total sales is; ', Total_sales, 'and the total pay is; ', Total_pay )
```
